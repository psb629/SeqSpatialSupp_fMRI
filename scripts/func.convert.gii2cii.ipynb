```python
for sn in tqdm(list_sn):
    subj = 'S'+sn

    fname = join(dir_surf,'glm_%1d'%glm,subj,'%s.%s.glm_%1d.beta.func.gii'%(subj,hem,glm))
    gii = nb.load(fname)
    
    N = len(gii.darrays)
    
    dataset = np.ones((N,32492)) * np.nan
    for ii, darray in enumerate(gii.darrays):
        data = darray.data
        dataset[ii] = data
    
    simage.save_surf2cifti(
        data=dataset,
        dir_output=join(dir_surf,'glm_%1d'%glm,subj),
        label_axis=['beta_%04d'%(nn+1) for nn in range(N)],
        prefix='%s.%s.glm_%1d.beta'%(subj,hem,glm),
        type_='dscalar'
    )
```
